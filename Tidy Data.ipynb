```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('./data/movie.csv')
movies.head()
```

```python
movies_new = movies.loc[:, ['title', 'actor1', 'actor2', 'actor3']].copy()
movies_new.head()
```

```python
melt = movies_new.melt(id_vars='title', var_name='level', value_name='name').sort_values(by='title')
melt.head()
```

```python
melt['name'].value_counts()
```

```python
a1 = movies['actor1'].dropna().value_counts()
a2 = movies['actor2'].dropna().value_counts()
a3 = movies['actor3'].dropna().value_counts()
```

```python
a1['Robert De Niro']
```

```python
a2['Robert De Niro']
```

```python
(a1 + a2 + a3).sort_values(ascending=False)
```

```python
a1.add(a2, fill_value=0).add(a3, fill_value=0).sort_values(ascending=False)
```
