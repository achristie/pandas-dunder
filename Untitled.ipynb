```python
flights = pd.read_csv('./data/flights.csv', parse_dates=['date'])
flights.insert(1, 'day_of_week', flights['date'].dt.day_name())
flights.insert(2, 'month', flights['date'].dt.month_name())
flights.head()
```

```python
pt = flights.pivot_table(index='airline', columns='day_of_week', aggfunc='mean', values='carrier_delay')
pt
```

```python
pt.style.highlight_max(axis='columns', color='red')
```

```python
flights.pivot_table(columns='airline', index='origin', aggfunc='sum', values='cancelled', margins=True)
```

```python
flights.head()
```

```python
pt = flights.pivot_table(index='airline', columns='month', aggfunc='sum', values='distance')
pt
```

```python
pt.style.highlight_max(axis='columns').format("{:,.0f}")
```

```python
pt = flights.pivot_table(index='month', columns='day_of_week', values='airline', aggfunc='size')
pt
```
